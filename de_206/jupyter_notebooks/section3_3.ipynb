# Section 3.3 Problem #24

Consider the initial value problem

$5u'' + 2u' + 7u = 0, \ \ u(0) = 2, \ \ u'(0) = 1$

(a) Find the solution u(t) of this problem.

```python
from manim import *
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np 
```

```python


u0 = 2  
uprime0 = 1  

def system(t, y):
    u, v = y
    du_dt = v
    dv_dt = (-2*v - 7*u)/5
    return [du_dt, dv_dt]

t_span = (0, 10)
t_eval = np.linspace(*t_span, 500)

sol = solve_ivp(system, t_span, [u0, uprime0], t_eval=t_eval)

plt.plot(sol.t, sol.y[0], label="u(t)")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.title("Static Solution  of the IVP")
plt.legend()
plt.show()
```

```python


class SolveIVPPlot(Scene):
    def construct(self):
        u0 = 2            # <-- Change the initial condition u(0)
        uprime0 = 1       # <-- Change the initial condition u'(0)

        def system(t, y):
            u, v = y
            du_dt = v
            dv_dt = (-2*v - 7*u)/5
            return [du_dt, dv_dt]

        t_span = (0, 10)
        t_eval = np.linspace(*t_span, 500)
        sol = solve_ivp(system, t_span, [u0, uprime0], t_eval=t_eval)

        ts = sol.t
        us = sol.y[0]

        ax = Axes(
            x_range=[0, 10, 2],
            y_range=[-3, 3, 1],
            axis_config={"color": WHITE},
            tips=False,
        )
        labels = ax.get_axis_labels(x_label="t", y_label="u(t)")

        points = [ax.c2p(t, u) for t, u in zip(ts, us)]
        graph = VMobject(color=YELLOW)
        graph.set_points_smoothly(points)

        self.play(Create(ax), Write(labels))
        self.wait(0.5)
        self.play(Create(graph), run_time=3)
        self.wait(0.5)

        initial_dot = Dot(ax.c2p(ts[0], us[0]), color=RED)
        initial_value = MathTex(f"u(0)={u0}").next_to(initial_dot, RIGHT * 4)
        self.play(FadeIn(initial_dot), Write(initial_value))
        self.wait(0.5)

        moving_dot = Dot(color=YELLOW).move_to(ax.c2p(ts[0], us[0]))
        self.add(moving_dot)
        def update_dot(mob, alpha):
            idx = int(alpha * (len(ts)-1))
            mob.move_to(ax.c2p(ts[idx], us[idx]))
        self.play(UpdateFromAlphaFunc(moving_dot, update_dot), run_time=3)
        self.wait(7)
```

(b) Find the smallest $T$ such that, for all $t > T,  \ \ \ |u(t)| \leq 0.1$. 

From the graph in part (a), we can see that the graph is oscillating, but approaches zero. It seems that for $t$ around $15$, the function values are between $-0.1$ and $0.1$, but this is a really rough visual guess.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# The ODE as a system
def odefunc(t, y):
    u, v = y
    du_dt = v
    dv_dt = (-2*v - 7*u)/5
    return [du_dt, dv_dt]

# Initial conditions
u0 = 2
v0 = 1
y0 = [u0, v0]

# Solve ODE on t = 0 to 30
t_span = (0, 30)
t_eval = np.linspace(*t_span, 1000)
sol = solve_ivp(odefunc, t_span, y0, t_eval=t_eval)
tpts = sol.t
upts = sol.y[0]

# Find the smallest T so that |u(t)| < 0.1 for all t > T
for i in range(len(upts)-1, -1, -1):
    if abs(upts[i]) > 0.1:
        T = tpts[i+1] if i+1 < len(tpts) else tpts[-1]
        print(f"For t > {T:.4f}, |u(t)| <= 0.1")
        break

# Plot
plt.plot(tpts, upts, label='u(t)')
plt.axhline(0.1, color='r', linestyle='--', label='|u(t)| = 0.1')
plt.axhline(-0.1, color='r', linestyle='--')
plt.axvline(T, color='g', linestyle=':', label=f'T  {T:.2f}')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.title('u(t) and the bound |u(t)| <= 0.1')
plt.legend()
plt.show()
```

From the above calculations we can see that, for all $t > 14.515$, we have $-0.1 < u(t) < 0.1$.
